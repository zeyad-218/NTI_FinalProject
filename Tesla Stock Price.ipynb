```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt 
```

```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
```

```python
tesla_data.head()
```

```python
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'], utc=True).dt.tz_localize(None)
```

```python
def make_graph(tesla_data, Tesla):
    plt.figure(figsize=(12, 6))
    plt.plot(tesla_data['Date'], tesla_data['Close'], linestyle='-', color='b')
    plt.title(f'{Tesla} Stock Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

make_graph(tesla_data, 'Tesla')
```
