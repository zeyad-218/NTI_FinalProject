```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

Tesla Model_X

```python
url_modelX = "https://www.goodcarbadcar.net/tesla-model-x-sales-figures-usa-canada/"
tesla_modelX = requests.get(url_modelX).text
tesla_modelX
```

```python
soup_modelX = BeautifulSoup(tesla_modelX,"html.parser")
soup_modelX
```

```python
table_modelX = soup_modelX.find('table')
table_modelX
```

```python
headers = [header.text.strip() for header in table_modelX.find_all('th')]
rows = []
for row in table_modelX.find_all('tr'):
    columns = row.find_all('td')
    row_data = [col.text.strip() for col in columns]
    if row_data: 
        rows.append(row_data)

df_modelX = pd.DataFrame(rows, columns=headers)
print(df_modelX)
```

```python
output_path = '/Users/zeyadosama/Desktop/Me/Training/NTI/Project/tesla_modelX_USAsales.xlsx'
df_modelX.to_excel(output_path, index=False)
```
